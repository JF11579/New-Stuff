```
condos2<- condos%>% group_by(Town) %>% mutate(town_Freq = n())
```

```
ggplot(condos2 ,aes(x=reorder(Town, town_Freq)))+
      geom_bar(color='skyblue',fill='steelblue',stat="count")+
      coord_flip()
```

```
ggplot(condos2 ,aes(x=reorder(Town, town_Freq)))+
      geom_bar(color='skyblue',fill='steelblue',stat="count",width=0.5)+
      coord_flip()+
      theme(text= element_text(size=25))
```
