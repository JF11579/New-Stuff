# Data Import

```r
# Units are thousands of people.
# Seasonally adjusted.
# Non-farm workers
data<- read.csv('/content/PAYEMS.csv')
head(data)
```

```r
tail(data)
```

# Pre-process data

```r
# convert DATE to date type

data$DATE <- as_date(data$DATE)
```

```r
# Let us look only at data since 2000
Labor_monthly <- data %>%
              filter(DATE >= as.Date("2000-01-01"))
```
